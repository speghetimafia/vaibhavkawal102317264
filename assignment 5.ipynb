Q1

```python
import numpy as np
import matplotlib.pyplot as plt

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
sum_all = np.sum(gfg)

# ii. Sum of all elements row-wise
sum_rows = np.sum(gfg, axis=1)

# iii. Sum of all elements column-wise
sum_cols = np.sum(gfg, axis=0)

print("Q1 Solutions:")
print("Sum of all elements:", sum_all)
print("Sum row-wise:", sum_rows)
print("Sum column-wise:", sum_cols)
```

Q2 (a)

```python

array = np.array([10, 52, 62, 16, 16, 54, 453])

# i. Sorted array
sorted_array = np.sort(array)

# ii. Indices of sorted array
sorted_indices = np.argsort(array)

# iii. 4 smallest elements
smallest_elements = np.sort(array)[:4]

# iv. 5 largest elements
largest_elements = np.sort(array)[-5:]

print("\nQ2(a) Solutions:")
print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 Smallest elements:", smallest_elements)
print("5 Largest elements:", largest_elements)
```

Q2 (b)

```python
array_b = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

# i. Integer elements only
int_elements = array_b[array_b == array_b.astype(int)]

# ii. Float elements only
float_elements = array_b[array_b != array_b.astype(int)]

print("\nQ2(b) Solutions:")
print("Integer elements:", int_elements)
print("Float elements:", float_elements)
```

Q3

```python
X = 86 + 75  # Replace with your initials' ASCII sum
sales = np.array([X, X+50, X+100, X+150, X+200])

tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 - tax_rate)

# Applying discounts
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)

# Expanding sales for multiple weeks
sales_matrix = np.vstack([sales, sales * 1.02, sales * 1.04])

print("\nQ3 Solutions:")
print("Sales array:", sales)
print("Tax rate:", tax_rate)
print("Sales after tax:", sales_after_tax)
print("Discounted sales:", discounted_sales)
print("Expanded sales data:", sales_matrix)
```
